```python
# Equation for a closed-open tube
# a is the first segment of the tube
# c is the last segment of the tube
# n is the formant number
# returns the frequencies for the specified segments of tubes per formant number
def closedOpen(a, c, n):
    newA = ((2*n - 1)*35000)/4*(a*.1)
    newC = ((2*n - 1)*35000)/4*(c*.1)
    
    return newA, newC
```

```python
# Equation for a closed-closed tube
# b is the middle segment of the tube
# n is the formant number
# returns the frequency for the specified segment per formant number
def closedClosed(b, n):
    newB = (n*35000)/(2*(b*.1))
    
    return newB
```

```python
# Vocal Tract Length
vtl = 175 # The vocal tract is 175 mm long; split on every mm cut possible to get three tubes

F1s = list()
F2s = list()
measurements = list()

# Split by 1 mm increments
for x in range(vtl):
    # Has to be split into 3 tubes
    if x != 0:
        y = vtl - x
        for z in range(y):
            # Split into 3 tubes
            if z != 0:
                # Calculate lengths for all three tubes
                a = x
                b = vtl - x - z
                c = vtl - x - b
                # Make sure that the three tubes add back up to the appropriate length
                assert(a+b+c == vtl),"Your vocal tract is the wrong length!"
                
                # Calculate appropriate frequencies for F1 and F2
                aF1, cF1 = closedOpen(a, c, 1)
                aF2, cF2 = closedOpen(a, c, 2)
                bF1 = closedClosed(b, 1)
                bF2 = closedClosed(b, 2)
                
                # Track triples of formants for each tube cut
                # One 3-tuple contains all of the formant frequencies for that particular cut of the vocal tract
                F1s.append((aF1, bF1, cF1))
                F2s.append((aF2, bF2, cF2))
                # Maintain an ordered list of cut points
                measurements.append((a, b, c))
```

```python
index = list() # Make a list to keep track of any and all frequencies below a given threshold (500 Hz)
lowest = 10000.0 # Arbitrarily large number
lowestCC = 20000.0 # Arbitrarily large number for closedClosed tube model

# Iterate through all triples of formant frequencies for F1
for triple in F1s:
    Q = False # Whether or not one of these will be lower than F1 for schwa in a 17.5cm vocal tract
    # Iterate through all the frequencies in the triple
    for item in triple:
        if item < 500:
            Q = True
        # Find the lowest frequency available to the data set
        if item < lowest:
            lowest = item
    if Q == True:
        index.append(F1s.index(triple))
print('The lowest frequency represented is', lowest)
```
